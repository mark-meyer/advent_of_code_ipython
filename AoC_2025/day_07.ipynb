```python
test_input=""".......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
...............""".split("\n")
```

```python
def parse_input(lines):
    start = lines[0].index("S")
    return start, [list(line) for line in lines]
        
```

```python
def part_one(start, lines):
    lines[1][start] = '|'
    beam_locations = set([start])
    splits = 0
    for row in lines[1:]:
        next_beam_locations = set()
        for beam in beam_locations:
            if row[beam] == '^':
                splits +=1
                next_beam_locations.add(beam-1)
                next_beam_locations.add(beam+1)
            else:
                next_beam_locations.add(beam)
        beam_locations = next_beam_locations
    return splits
```

```python
start, lines = parse_input(test_input)

part_one(start, lines)
```

```python
with open("input_files/day_07.txt") as f:
    raw_input = f.read().splitlines()
    start, lines = parse_input(raw_input)

part_one(start, lines)
```

```python
from functools import cache

def part_two(start, lines):
    @cache
    def count_states(row, col):
        if row + 1 == len(lines):
            return 1
    
        if lines[row+1][col] == '^':
            return count_states(row+1, col-1) + count_states(row+1, col+1)
        else:
            return count_states(row+1, col)
            
    return count_states(1, start) 

start, lines = parse_input(test_input)
part_two(start, lines)
```

```python
with open("input_files/day_07.txt") as f:
    raw_input = f.read().splitlines()
    start, lines = parse_input(raw_input)
part_two(start, lines)
```
