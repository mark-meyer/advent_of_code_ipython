```python
test_input="""987654321111111
811111111111119
234234234234278
818181911112111
"""
```

```python
def parse_input(s):
    lines = s.split()
    return [[int(c) for c in line] for line in lines]

test_input = parse_input(test_input)
test_input
```

```python
with open("input_files/day_03.txt") as f:
    raw_data = f.read()

data = parse_input(raw_data)
```

```python
def high_joltage(line, length):
    digits = 0
    l_index = 0
    for r_index in range(1-length, 1):
        segment = line[l_index:r_index] if r_index < 0  else line[l_index:]
        l_index, n = max(enumerate(segment, start=l_index+1), key=lambda line: (line[1], -line[0]))
        digits += 10 ** abs(r_index) * n
    return digits
```

```python
def part_one(data):
    return sum(high_joltage(line, 2) for line in data)

part_one(test_input)
```

```python
part_one(data)
```

```python
def part_two(data):
    return sum(high_joltage(line, 12) for line in data)
        
part_two(test_input)
```

```python
part_two(data)
```
