```python
def parse_input(s: str):
    ranges, ingredients = s.split("\n\n")
    ranges = sorted(tuple(int(n) for n in range.split('-')) 
              for range in ranges.split('\n'))
    ingredients = [int(n) for n in ingredients.split('\n')]
    return ranges, ingredients 

ranges, ingredients = parse_input(test_input)
ranges
ranges_rev = sorted(ranges, key=lambda t: t[1])
ingredients, ranges
```

## Part One

Was expecting part 2 to be harder, to the binary search for ranges it probably overkill.

This finds all ranges where the start is smaller than n and end is larger than n.

```python
import bisect

def get_ranges(n, ranges, ranges_rev):
    small = bisect.bisect(ranges, n, key=lambda t: t[0])
    large = bisect.bisect_left(ranges_rev, n, key=lambda t: t[1])
    return set(ranges[:small]) & set(ranges[large:])
    
def part_one(s):
    ranges, ingredients = parse_input(s)
    ranges_rev = sorted(ranges, key=lambda t: t[1])
    
    total = 0
    for i in ingredients:
        found_ranges = get_ranges(i, ranges, ranges_rev)
        if found_ranges:
            assert (i >= r[0] and i <= r[1] for r in found_ranges)
            total += 1
    return total


part_one(test_input)
```

## Part Two

Sort ranges by start of range. Then iterate and merge based on overlap.

```python
def part_two(ranges):
    current = ranges[0]
    total = 0
    for r in ranges[1:]:
        if r[0] <= current[1]:
            if r[1] > current[1]:
                current = (current[0], r[1])
            # otherwise range is completely included in previous
        else:
            total += current[1] - current[0] + 1
            current = r
    total += current[1] - current[0] + 1
            
        
    return total

ranges, _ = parse_input(test_input)
part_two(ranges)
```

```python
ranges, _ = parse_input(raw_input)

part_two(ranges)
```
